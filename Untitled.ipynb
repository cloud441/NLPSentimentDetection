## Data set

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data()
word_index = keras.datasets.imdb.get_word_index()
inverted_word_index = dict((i, word) for (word, i) in word_index.items())
```

# Simple version

```python
classes = [0, 1] # 0 is the class for negatives review and 1 for positives
print()

def train_naive_bayes(dataset, classes, vocabulary) -> (int,int) :
    
    number_documents = len(x_train)
    y_train_occurences = np.bincount(y_train)
    
    loglikehood = {}
    logprior = []
    
    #computing each class
    for c in classes:
        number_documents_c = y_train_occurences[c]
        
        logprior.append(math.log(number_documents_c / number_documents))
        
        #Fill the bag of word :
        bag_of_word = []
        for i, word in enumerate(x_train):
            #word is in the class
            if y_train[i] == c:
                bag_of_word.append(word)
        
        for word in vocabulary:
            count_word_c = bag_of_word.count(word)
            loglikehood[(c,word)] = math.log((count_word_c + 1)/(len(bag_of_word) + 1))
    
    return (logprior, loglikehood)

(logprior, loglikehood) = train_naive_bayes(5, classes, inverted_word_index)
```
